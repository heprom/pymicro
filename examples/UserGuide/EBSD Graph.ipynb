```python
import sys
sys.path.append('../..')
from pymicro.crystal.microstructure import Microstructure
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
m = Microstructure('../data/Ti64-200x200_data.h5')
```

```python
m.get_all_grains()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(m.euler[:,:,0])
ax[0].axis('off')
ax[0].set_title('Orientation map')
ax[1].imshow(m.get_grain_map())
ax[1].axis('off')
ax[1].set_title('Grain IDs')

plt.show()
```
